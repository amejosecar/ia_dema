# Seaborn

Seaborn complementa a Mathplotlib, con una filosofía de hacer fácil lo menos fácil.  Incluye temas personalizados que nos permiten controlar la forma como se dibujan los gráficos.

En este sentido soluciona 2 problemas existentes en Matplotlib:

* Parámetros por defecto.
* Trabajar con DataFrames.

### Trabajando con colores

Parámetros:
* n_colors: Nº de colores de la paleta.  Por defecto = 6
* desat: % de desaturación para cada color.

Por defecto, tenemos 6 paletas:

* deep
* muted
* brigth
* pastel
* dark
* colorblind

Pero podemos crear nuestras propias paletas

```python
sns.color_palette(palette=['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'brown'], n_colors=7, desat= 1)
```

```python
sns.color_palette(palette=['red', 'blue', 'green'], n_colors=6, desat= 0.5)
```

```python
sns.color_palette(palette=['red', 'blue', 'green'], n_colors=3, desat= 0.25)
```

```python
paleta1 = sns.color_palette(palette=['red', 'blue', 'green'], n_colors=3, desat= 1)
```

```python
paleta1
```

```python
# Usamos la nueva paleta.
sns.set(context='notebook', style='darkgrid', palette=paleta1, font='sans-serif', font_scale=1, color_codes=False, rc=None)
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,x)
plt.show()
```

```python
# Usamos una paleta por defecto.
sns.set(context='notebook', style='darkgrid', palette='colorblind', font='sans-serif', font_scale=1, color_codes=False, rc=None)
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,x)
plt.show()
```

Si usamos despine() junto con set_style("white"), sólo se mostrarán 2 de los 4 marcos de la imagen (frente a ninguno si los usamos con otro set_style)

```python
sns.set(style="white", palette=paleta1)
sns.set_style('white')
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,x)
sns.despine()
plt.show()
```

Podemos usar 3 tipos diferentes de paletas:

https://seaborn.pydata.org/tutorial/color_palettes.html

* qualitative.  Colores categóricos (rojo, verde, azul, ...
* sequential. Secuencia de diferentes tonalidades de un color
* diverging. Colores divergentes

Definiremos la paleta a usar con **palplot()**

```python
# qualitative.  1 paleta. tab10.  Si no se indica otra, es la que toma por defecto
paleta = sns.color_palette('tab10',8)
sns.palplot(paleta)
plt.show()
```

```python
# sequential
# disponemos de 4 paletas secuenciales por defecto: rocket, mako, flare, crest
paleta = sns.color_palette('viridis',10)
sns.palplot(paleta)
plt.show()
```

```python
# divergent.
# Se basa en representar 2 colores
paleta = sns.color_palette("light:b", 14) # Paleta con 7 colores empezando por el marrón (br) y terminando en el verde oscuro (BG)
sns.palplot(paleta)
plt.show()
```

Como hemos visto antes, a través de **set_style()** definimos las opciones por defecto, incluida la paleta a usar.

### Donuts concéntricos

```python
# Datos para el primer gráfico de donut (exterior)
sizes_exterior = [30, 20, 50]  # Tamaños de las porciones
labels_exterior = ['A', 'B', 'C']  # Etiquetas de las porciones
colors_exterior = ['#ff9999', '#66b3ff', '#99ff99']  # Colores de las porciones

# Datos para el segundo gráfico de donut (interior)
sizes_interior = [10, 40]  # Tamaños de las porciones
labels_interior = ['X', 'Y']  # Etiquetas de las porciones
colors_interior = ['#ffcc99', '#c2c2f0']  # Colores de las porciones

# Crear el gráfico de donut exterior
fig, ax = plt.subplots()

# Dibujar el gráfico de donut exterior
wedges_exterior, texts_exterior, autotexts_exterior = ax.pie(sizes_exterior, labels=labels_exterior, colors=colors_exterior,
                                                             autopct='%1.1f%%', startangle=90, pctdistance=0.85)

# Hacer que el gráfico de donut sea un anillo (donut)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)

# Dibujar el gráfico de donut interior
wedges_interior, texts_interior, autotexts_interior = ax.pie(sizes_interior, labels=labels_interior, colors=colors_interior,
                                                             autopct='%1.1f%%', startangle=90, pctdistance=0.75, radius=0.69)

# Hacer que el gráfico de donut interior sea un anillo (donut)
centre_circle_interior = plt.Circle((0, 0), 0.40, fc='white')
fig.gca().add_artist(centre_circle_interior)

# Asegurarse de que el gráfico sea un círculo perfecto
ax.axis('equal')

# Mostrar el gráfico
plt.show()
```
