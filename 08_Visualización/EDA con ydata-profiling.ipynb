```python
g = sns.catplot(
    data=datos, x="Sex", y="Survived", col="Pclass",
    kind="bar", height=4, aspect=.6,
)
g.set_axis_labels("Sexo", "Survival Rate")
g.set_xticklabels(["male", "female"])
g.set_titles("{col_name} {col_var}")
g.set(ylim=(0, 1))
g.despine(left=True)
```
