```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap
numeric_df = df2.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

# Crear una máscara para enmascarar la diagonal principal y los valores por encima
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask)
plt.title('Correlation Heatmap')
plt.show()
```

## 14. Crear un nuevo Data Frame con las 10 variables más correladas.


```python
corr_con_SP = df2_num[df2_num.corr()["SalePrice"].sort_values(ascending=False)[:10].index]
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation heatmap
def correlacion(df2):
    numeric_df = df2.select_dtypes(include=[np.number])
    corr_matrix = numeric_df.corr()

    # Crear una máscara para enmascarar la diagonal principal y los valores por encima
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

    plt.figure(figsize=(12, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask=mask)
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
correlacion(corr_con_SP)
```
