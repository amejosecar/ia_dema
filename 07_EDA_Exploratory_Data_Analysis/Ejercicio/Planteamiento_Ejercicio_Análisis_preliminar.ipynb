## 5. Sustituir los “Si” por “Crisis” y los “No” por “Burbuja”.
    - Se puede hace de 2 maneras
      - .replace
      - .map
      - .cat.rename_categories


```python
df1["Crisis"] = df1["Crisis"].replace({"Si": "Crisis", "No": "Burbuja"})
df1.head(10)
```

```python
df1["Crisis"] = df1["Crisis"].map({"Si": "Crisis", "No": "Burbuja"})
df1.head(10)
```

```python
df1["Crisis"] = df1["Crisis"].astype("category")
```

```python

df1["Crisis"].cat.rename_categories({"Si": "Crisis", "No": "Burbuja"})

df1.head(10)
```

```python
Total_por_años = df1.groupby('YrSold')['SalePrice'].sum()
```

```python
# 2. Graficamos con pandas solo (bar plot sencillo)
ax = Total_por_años.plot(
    kind='bar',
    figsize=(10, 6),
    color='#4C72B0',
    title='Suma Total de Precios de Venta por Año'
)
ax.set_xlabel('Año de Venta')
ax.set_ylabel('Suma de SalePrice (€)')
ax.set_xticklabels(Total_por_años.index, rotation=0)
plt.tight_layout()
plt.show()
```
