## Tipos de gráficos

* **Plot** Pinta líneas o marcadores en los ejes.
* **Bar** Para pintar gráficos de barras.
* **Barh** Pinta un gráfico de barras horizontal.
* **Boxplot** Pinta gráficos de barrasMake a box and whisker plot.
* **Hist** Pinta un histograma.
* **hist2d** Crea un histograma en 2D.
* **Pie** Pinta un gráfico de tarta.
* **Polar** Pinta un plot en base a coordenadas polares.
* **Scatter** Pinta un gráfico de dispersión (x vs y).
* **Stackplot** Pinta un gráfico de áreas apiladas.
* **Stem** Crea un gráfico de tallos y hojas.
* **Step** Crea un gráfico (mercado de valores).
* **Quiver** Crea un campo de flechas en un gráfico 2-D.

### Funciones de los ejes

* **Axes** Añade ejes a la figura.
* **Text** Añade texto a los ejes.
* **Title** Añade título a la gráfica.
* **Xlabel** Añade título al eje X.
* **Xlim** Obtiene o establece los límites del eje x.
* **Xscale** Establece el escalado del eje x.
* **Xticks**
* **Ylabel** Añade título al eje Y.
* **Ylim** Obtiene o establece los límites del eje y.
* **Yscale** Establece el escalado del eje y.
* **Yticks**

### Funciones de los gráficos (figuras)

* **Figtext** Añade texto a la figura.
* **Figure** Crea una nueva figura.
* **Show** Muestra una figura (gráfico).
* **Savefig** Guarda el gráfico actual.
* **Close** Cerrar el gráfico actual.

## Objeto `figure` y sus ejes

```python
# Recuperamos la función senoidal. De 0 a 4π
angulo = np.arange(0, math.pi*4, 0.05) # Rango de datos para pintar el eje X
seno = np.sin(angulo)
```

Vamos a definir una figura y añadirle unos ejes, que nos sirven para dar dimensiones al dibujo.

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html


La línea de código "ax = fig.add_axes([0,0,2,1])" se utiliza para agregar un nuevo conjunto de ejes (axes) a una figura existente.

- El primer parámetro, 0, representa la posición del eje en el eje x, donde 0 significa que el eje comienza desde el **borde izquierdo** de la figura.
- El segundo parámetro, 0, representa la posición del eje en el eje y, donde 0 significa que el eje comienza desde el **borde inferior** de la figura.
- El tercer parámetro, 2, representa el ancho del eje en unidades relativas. En este caso, el eje se extiende dos veces más allá de la figura en el eje x.
- El cuarto parámetro, 1, representa la altura del eje en unidades relativas. En este caso, el eje tiene la misma altura que la figura en el eje y.

```python
# Creamos el objeto figura
fig = plt.figure()
"""
Añadimos los ejes.
add_axes espera una lista con 4 elementos correspondientes a los elementos
    izquierda, abajo, ancho y altura de los ejes.
"""
ax = fig.add_axes([0,0,2,1])
# Definir el título y etiquetas de los ejes en la figura.
ax.set_title("Onda senoidal")
ax.set_xlabel('Ángulo')
ax.set_ylabel('Seno')
ax.plot(angulo, seno)
# Podemos añadir una leyenda al gráfico.
ax.legend(labels=('Seno del ángulo',), borderpad= 0.5, loc='best')
plt.show()
```

### Subplots
A través de esta opción podremos pintar diferentes gráficas en el mismo 'lienzo'

**plt.subplot(subplot(nrows, ncols, index))**

En este sentido plt.subplot([1,2,1]), define un lienzo con 1 fila, 2 columnas y el gráfico a pintar ocupará la primera posición.
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html


```python
# Al pintar una línea en un gráfico de manera implícita estamos creando un subplot (111)
plt.plot([1,2,8,4])
```

```python
# Creamos ahora un subplot compuesto de una malla con 2 filas y 1 columna
plt.subplot(211)
plt.plot(range(12))
plt.subplot(212, facecolor='#dc1be3') # generamos el segundo plot, con un fondo morado.
plt.plot(range(12), '#1be3cf')
```

```python
# Podemos superponer subplots en otros
fig=plt.figure()
# Este 'ax1' lo ocupa todo (una fila, una columna, índice 1)
ax1=fig.add_subplot(111)
# Mientras 'ax2' ocupa la izquierda (una fila, tres columnas, ocupa el índice 1)
ax2=fig.add_subplot(131, facecolor='y')
# 'ax3' ocupa la esquina derecha (dos filas, tres columnas, ocupa el índice 6)
ax3=fig.add_subplot(236, facecolor='r')
ax1.plot([1,2,5])
ax2.plot([1,2,2])
ax3.plot([1,2,3])
plt.show()

# Vemos que al definir un subplot diferente en ax1 y ax2, las 2 gráficas se superponen.
```

```python
# También nos puede interesar, no superponer gráficas, sino contraponerlas...
fig, a = plt.subplots(3,2) # 3 filas y 2 columnas
fig.subplots_adjust(hspace=0.5, wspace=0.5)
x=np.arange(0.1,5,0.10)
a[0][0].plot(x,x*x)
a[0][0].set_title('cuadrado')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('raíz cuadrada')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
a[2][0].plot(x,np.exp(x))
a[2][0].set_title('nuevo exp')
a[2][1].plot(x,np.log10(x))
a[2][1].set_title('nuevolog')
plt.show()
```

```python
# Siguiendo la metodología previa, podemos generar gráficas de diferente tamaño
a1= plt.subplot2grid((4,2),(0,0), colspan=2) # Pinta a1 en una malla de 3x3, en la posición 0,0 y compuesta de 2 columnas (2 celdas en la malla)
a2= plt.subplot2grid((4,2),(1,1), rowspan=2) # Pinta a2 en una malla de 3x3, en la posición 0,2 y compuesta de 3 filas (3 celdas en la malla)
a3= plt.subplot2grid((4,2),(1,0), rowspan=3) # Pinta a3 en una malla de 3x3, en la posición 1,0 y compuesta de 2 filas y 2 columnas (4 celdas en la malla)
x=np.arange(1,10, 0.2)
a1.plot(x, np.exp(x))
a1.set_title('exp')
a2.plot(x, x*x)
a2.set_title('cuadrado')
a3.plot(x, np.log(x))
a3.set_title('log')
plt.tight_layout()
plt.show()
```

### Cuadrícula en la gráfica

```python
# Por claridad, nos puede interesar mostrar la cuadrícula en el gráfico

fig, axes = plt.subplots(1,3, figsize=(12,4))
x=np.arange(-4, 4, 0.1)
axes[0].plot(x, np.tanh(x), 'g', lw=5) # Modificado grosor de la línea del gráfico.
axes[0].grid(True)
axes[0].set_title('cuadrícula por defecto')
axes[1].plot(x, 1 / (1 + np.exp(-x)), 'r')
axes[1].grid(color='b', ls='--', lw=0.5)
axes[1].set_title('cuadrícula personalizada')
axes[2].plot(x,np.maximum(0, x))
axes[2].set_title('sin cuadrícula')
fig.tight_layout()
plt.show()
```

## Pintando unas gráficas

### Gráficos de barras

**ax.bar(x, height, width, bottom, align)**

* x: datos x a representar (en formato array)
* height: datos y a representar (en formato array)
* width: ancho de las barras (en formato array)
* bottom: define el origen de datos para el eje Y
* align: alineación de los datos,'center' o 'edge', respecto de los marcadores (ticks)

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
prod=['TV', 'Netflix', 'Pluto', 'Rakuten', 'Otros']
usuarios=[23, 17, 35, 29, 12]
ax.bar(prod, usuarios, align="edge", width=0.8)
plt.show()
```

¡Ojo! Porque si definimos bottom, se desplaza el eje Y.  En el ejemplo, Y=23, tras el desplazamiento se representa como Y=123

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
prod=['TV', 'Netflix', 'Pluto', 'Rakuten', 'Otros']
usuarios=[23,17,35,29,12]
ax.bar(prod,usuarios, bottom=100, align="center", width=0.5)
plt.show()
```

## Ejemplo con el campo vectorial de dos electrones

Para graficar el campo vectorial generado por dos electrones en las posiciones $(-1, -1)$ y $(1, 1)$, puedes calcular el campo eléctrico en cada punto de la cuadrícula utilizando la ley de Coulomb. El campo eléctrico generado por una carga puntual $q$ en la posición ($x_0$, $y_0$) en un punto ($x$, $y$) se calcula como:

$\mathbf{E} = \frac{k \cdot q}{r^2} \cdot \hat{r}$

Donde:
- $k$ es la constante de Coulomb (en unidades apropiadas).
- $r$ es la distancia entre la carga y el punto \((x, y)\).
- $\hat{r}$ es el vector unitario en la dirección de la carga al punto.

Para dos cargas, el campo eléctrico total es la suma de los campos generados por cada carga.

### Código para graficar el campo vectorial:



```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la constante de Coulomb (en unidades apropiadas)
k = 1

# Posiciones de las cargas
q1 = 1  # Carga en (-1, -1)
q2 = 1  # Carga en (1, 1)
pos1 = np.array([-1, -1])
pos2 = np.array([1, 1])

# Crear la cuadrícula
x, y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))

# Calcular el campo eléctrico en cada punto de la cuadrícula
Ex, Ey = np.zeros_like(x), np.zeros_like(y)

for i in range(x.shape[0]):
    for j in range(x.shape[1]):
        # Punto actual (x, y)
        r = np.array([x[i, j], y[i, j]])

        # Campo generado por la carga en (-1, -1)
        r1 = r - pos1
        r1_mag = np.linalg.norm(r1)
        if r1_mag > 0:  # Evitar división por cero
            E1 = k * q1 / r1_mag**2
            Ex[i, j] += E1 * (r1[0] / r1_mag)
            Ey[i, j] += E1 * (r1[1] / r1_mag)

        # Campo generado por la carga en (1, 1)
        r2 = r - pos2
        r2_mag = np.linalg.norm(r2)
        if r2_mag > 0:  # Evitar división por cero
            E2 = k * q2 / r2_mag**2
            Ex[i, j] += E2 * (r2[0] / r2_mag)
            Ey[i, j] += E2 * (r2[1] / r2_mag)

# Graficar el campo vectorial
plt.figure(figsize=(8, 8))
plt.quiver(x, y, Ex, Ey, scale=20, scale_units='xy')
plt.scatter([-1, 1], [-1, 1], color='red', s=100, label='Cargas')  # Marcar las posiciones de las cargas
plt.title('Campo Vectorial Eléctrico')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

### Explicación del código:
1. **Constante de Coulomb**: Se define como \(k = 1\) (puedes ajustarla según las unidades que uses).
2. **Posiciones de las cargas**: Las cargas están en \((-1, -1)\) y \((1, 1)\).
3. **Cuadrícula**: Se crea una cuadrícula con `np.meshgrid` para evaluar el campo en cada punto.
4. **Cálculo del campo eléctrico**:
   - Para cada punto \((x, y)\), se calcula la distancia a cada carga.
   - Se calcula el campo eléctrico generado por cada carga utilizando la ley de Coulomb.
   - Se suman los campos generados por ambas cargas para obtener el campo total.
5. **Gráfico**:
   - Se usa `plt.quiver` para graficar los vectores del campo eléctrico.
   - Se marcan las posiciones de las cargas con puntos rojos.

### Resultado:
El gráfico mostrará el campo vectorial eléctrico generado por las dos cargas en el plano. Los vectores apuntarán hacia afuera de las cargas si estas son positivas (como en este caso).

```python
# Otro ejemplo, con 2 plots

# Definimos la malla para los plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize =(14, 8))

# Primer subplot
# Generación de los vectores
x = np.arange(0, 2.2, 0.2)
y = np.arange(0, 2.2, 0.2)
X, Y = np.meshgrid(x, y)
u = np.cos(X)*Y
v = np.sin(y)*Y
n = -2

# Definición del color
color = np.sqrt(((v-n)/2)*2 + ((u-n)/2)*2)

# Creación del plot
ax1.quiver(X, Y, u, v, color, alpha = 0.8)
ax1.xaxis.set_ticks([])
ax1.yaxis.set_ticks([])
ax1.axis([-0.2, 2.3, -0.2, 2.3])
ax1.set_aspect('equal')
ax1.set_title('Función meshgrid')

# Segundo subplot
# Creación de los vectores
x = np.arange(-2, 2.2, 0.2)
y = np.arange(-2, 2.2, 0.2)
X, Y = np.meshgrid(x, y)
z = X * np.exp(-X**2 -Y**2)
dx, dy = np.gradient(z)
n = -2

# Defining color
color = np.sqrt(((dx-n)/2)*2 + ((dy-n)/2)*2)

# Creating plot
ax2.quiver(X, Y, dx, dy, color)
ax2.xaxis.set_ticks([])
ax2.yaxis.set_ticks([])
ax2.set_aspect('equal')
ax2.set_title('gradient')


# show figure
plt.tight_layout()
plt.show()
```
